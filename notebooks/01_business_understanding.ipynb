A ideia aqui não é apenas descrever o problema, mas traduzi-lo em uma formulação analítica clara. Essa etapa é fundamental dentro do fluxo do CRISP-DM, já que orienta todas as decisões seguintes, desde a análise exploratória até a modelagem.

<h2 style="color:#2E86C1; border-bottom: 2px solid #2E86C1; padding-bottom: 5px;">
1. Contexto do problema
</h2>

O case descreve uma empresa de e-commerce que passou por um crescimento acelerado. Com isso, aumentaram não só os volumes de pedidos e entregas, mas também a complexidade da operação como um todo.

Esse ganho de escala veio acompanhado de um efeito colateral importante: uma maior variabilidade na experiência do cliente, refletida no NPS.

Hoje, o NPS é coletado apenas no final da jornada. Na prática, isso significa que a empresa só descobre que algo deu errado quando já não há mais espaço para correção. O feedback chega tarde demais para qualquer ação preventiva.

Nesse contexto, o problema não é apenas medir satisfação, mas entender o que está por trás dela. Ou seja, quais fatores operacionais, como logística, prazo e atendimento, estão impactando a experiência do cliente ao longo da jornada.

A necessidade do negócio passa, então, por transformar dados operacionais, que já existem, em sinais antecipados de satisfação. A ideia é sair de um modelo reativo, em que o NPS é analisado apenas depois, para um modelo mais preditivo, que permita atuar antes da experiência se consolidar.

<h3 style="color:#2E86C1;">
1.1 O que é NPS
</h3>

O Net Promoter Score (NPS) é uma métrica utilizada para medir a satisfação e a lealdade do cliente a partir de uma única pergunta: qual a probabilidade de recomendar a empresa para outra pessoa.

Com base na resposta, os clientes são classificados em três grupos:

- **Promotores (9–10):** clientes satisfeitos, com maior probabilidade de recompra e recomendação  
- **Neutros (7–8):** clientes indiferentes, com menor nível de engajamento  
- **Detratores (0–6):** clientes insatisfeitos, com maior risco de churn e impacto negativo na reputação  

<h2 style="color:#2E86C1; border-bottom: 2px solid #2E86C1; padding-bottom: 5px;">
2. Entendimento do negócio
</h2>

<h3 style="color:#2E86C1;">
2.1 Qual problema de negócio está sendo resolvido?
</h3>

O problema de negócio está na dificuldade da empresa em antecipar experiências negativas ao longo da jornada de compra.

Hoje, a percepção do cliente só aparece no final do processo, via NPS. Isso faz com que a empresa acabe atuando de forma reativa: quando um cliente é classificado como detrator, o problema já aconteceu e, muitas vezes, não há mais oportunidade de corrigir a experiência.

Na prática, isso limita a capacidade de:

- prevenir falhas operacionais  
- priorizar casos mais críticos  
- agir no momento em que ainda seria possível impactar a experiência  

Dessa forma, o desafio não é apenas medir satisfação, mas entender o que leva a uma boa ou má experiência e, principalmente, conseguir usar esses sinais antes do fim da jornada.

<h3 style="color:#2E86C1;">
2.2 Por que o NPS é importante para um e-commerce?
</h3>

O NPS é relevante porque resume, em uma única métrica, a percepção final do cliente sobre toda a experiência de compra.

No e-commerce, essa experiência é formada por vários pontos de contato: preço, prazo de entrega, logística, atendimento, resolução de problemas e confiança na marca. O NPS funciona como uma síntese de tudo isso. Ele não mede uma etapa isolada, mas o resultado final da jornada.

Além disso, ele permite que a empresa:

- entenda se está gerando valor ou frustração para o cliente  
- compare desempenho entre diferentes perfis, regiões ou fluxos operacionais  
- conecte indicadores operacionais com a percepção do cliente  
- acompanhe se melhorias implementadas realmente impactam a experiência  

Por isso, o NPS não é apenas uma métrica de acompanhamento, mas um indicador importante para orientar decisões de negócio mais centradas no cliente.

<h3 style="color:#2E86C1;">
2.3 Quais áreas poderiam se beneficiar desses insights?
</h3>

Os insights gerados a partir da análise de NPS não ficam restritos a uma única área. Eles acabam impactando praticamente toda a operação.

**Logística**  
A logística é uma das principais responsáveis pela experiência no e-commerce. Aqui, os dados podem ajudar a:

- entender o impacto de atrasos, múltiplas tentativas de entrega e tempo total de entrega na satisfação  
- identificar regiões, transportadoras ou fluxos logísticos mais problemáticos  

**Atendimento ao cliente**  
O atendimento normalmente entra em cena quando algo já deu errado, então acaba sendo um bom termômetro de fricção. Os insights podem ser usados para:

- avaliar como número de contatos, reclamações e tempo de resolução influenciam o NPS  
- priorizar clientes com maior risco de se tornarem detratores  

**Produto e experiência do cliente**  
Nem todo problema está na operação. Parte pode estar no próprio desenho da jornada. Nesse caso, a análise ajuda a:

- identificar padrões de pedidos ou fluxos que geram mais atrito  
- orientar melhorias estruturais na experiência como um todo  

**Pricing / Comercial**  
A percepção de valor também influencia a satisfação. A análise pode ajudar a:

- entender se valor do pedido, frete, descontos e parcelamento influenciam o NPS  
- avaliar se o cliente sente que “valeu a pena” a compra  

**Estratégia e liderança**  
Em um nível mais estratégico, o valor está em transformar dado em ação:

- sair de uma leitura reativa para uma atuação mais preventiva  
- apoiar decisões baseadas em evidência para melhorar retenção, reputação e eficiência operacional  

No fim, o NPS funciona como uma camada que conecta operação e percepção do cliente. É isso que permite alinhar as diferentes áreas em torno da mesma métrica.

<h2 style="color:#2E86C1; border-bottom: 2px solid #2E86C1; padding-bottom: 5px;">
3. Como o NPS impacta o negócio
</h2>

<h3 style="color:#2E86C1;">
3.1 Recompra
</h3>

O NPS tende a estar diretamente relacionado à probabilidade de recompra. Clientes mais satisfeitos têm maior propensão a voltar a comprar, enquanto experiências negativas podem levar à redução da frequência, abandono ou migração para concorrentes.

No contexto de e-commerce, isso ganha ainda mais relevância, já que o custo de aquisição de novos clientes costuma ser alto. Melhorar a experiência de quem já comprou tende a ser mais eficiente do que depender apenas da entrada de novos clientes.

<h3 style="color:#2E86C1;">
3.2 Boca a boca
</h3>

O NPS também impacta o boca a boca. Clientes promotores tendem a recomendar a marca de forma espontânea, enquanto detratores podem compartilhar experiências negativas, muitas vezes em canais com grande alcance, como redes sociais e sites de avaliação.

Como a reputação digital influencia diretamente a decisão de compra no e-commerce, esse efeito pode amplificar tanto ganhos quanto perdas.

<h3 style="color:#2E86C1;">
3.3 Market share em e-commerce
</h3>

No médio e longo prazo, uma experiência mais consistente contribui para retenção, reputação e preferência de marca, o que pode ajudar a empresa a manter ou expandir seu market share.

Por outro lado, falhas recorrentes na experiência tendem a enfraquecer a competitividade, principalmente em mercados com baixa diferenciação e alta concorrência.

<h2 style="color:#2E86C1; border-bottom: 2px solid #2E86C1; padding-bottom: 5px;">
4. Indicadores de mercado que podem complementar a análise
</h2>

Embora o NPS seja uma métrica importante, ele não deve ser analisado de forma isolada. Para ter uma visão mais completa do negócio, faz sentido combiná-lo com outros indicadores operacionais e de mercado.

Alguns exemplos relevantes incluem:

- benchmark de NPS do setor, para entender se a empresa está acima ou abaixo do mercado  
- SLA logístico, comparando prazo prometido versus prazo realizado  
- taxa de atraso e taxa de reentrega  
- tempo médio de resolução no atendimento  
- volume de reclamações  
- taxa de recompra  
- taxa de cancelamento ou abandono  
- indicadores competitivos, como percepção de marca ou participação em segmentos-chave  

Esses indicadores ajudam a dar mais contexto ao NPS, conectando a percepção do cliente com a performance operacional e com a posição da empresa no mercado.

<h2 style="color:#2E86C1; border-bottom: 2px solid #2E86C1; padding-bottom: 5px;">
5. Definição da target
</h2>

Seguindo a lógica do CRISP-DM, depois de entender o problema de negócio, o próximo passo é traduzi-lo para um objetivo analítico claro.

Na prática, isso significa sair de uma dor mais ampla, como “melhorar a experiência do cliente”, e transformar isso em algo que possa ser medido e trabalhado com dados.

Em projetos supervisionados, essa etapa se resume à definição da variável-alvo (target), ou seja, qual fenômeno queremos prever ou explicar a partir do histórico disponível.

Essa definição é crítica, porque direciona todo o restante do projeto, desde a seleção de variáveis até a escolha do modelo e das métricas de avaliação.

<h3 style="color:#2E86C1;">
5.1 Qual variável representa a satisfação do cliente?
</h3>

A variável que representa diretamente a satisfação do cliente é **`nps_score`**.

Ela é a melhor candidata a target porque reflete, dentro do dataset, a avaliação final que o cliente fez da sua experiência de compra.

<h3 style="color:#2E86C1;">
5.2 Por que ela foi escolhida?
</h3>

A variável `nps_score` foi escolhida porque:

1. representa diretamente o fenômeno de interesse do negócio, que é a satisfação do cliente  
2. está alinhada ao problema central do case, que é a variabilidade da experiência entre diferentes clientes  
3. permite conectar fatores operacionais observáveis, como entrega, atendimento, reclamações e tempo de resolução, à percepção final  
4. pode ser usada tanto em análises descritivas quanto, posteriormente, em abordagens preditivas  

Ou seja, o `nps_score` funciona como o ponto de ligação mais direto entre a dor de negócio e a modelagem analítica.

<h3 style="color:#2E86C1;">
5.3 Em que momento da jornada essa informação é coletada?
</h3>

Essa informação é coletada **após o encerramento da jornada de compra**.

Isso é relevante porque mostra que o NPS é uma medida de resultado. Ele reflete a experiência completa, mas não funciona como um sinal antecipado ao longo do processo.

<h3 style="color:#2E86C1;">
5.4 Existe algum risco de usar essa variável de forma inadequada?
</h3>

Sim, existem alguns pontos de atenção importantes:

**1. Tratar o NPS como se fosse toda a experiência**  
O NPS resume a experiência em uma única nota, mas não captura todas as nuances da jornada do cliente.

**2. Ignorar o aspecto temporal**  
Como o NPS é coletado no final, é preciso ter cuidado para não usar, em um modelo preditivo, informações que só estariam disponíveis depois da experiência concluída. Isso geraria vazamento de informação (data leakage).

**3. Confundir correlação com causalidade**  
Uma variável estar associada ao NPS não significa, necessariamente, que ela causa satisfação ou insatisfação. A interpretação precisa levar em conta o contexto de negócio.

**4. Viés de resposta**  
Nem todos os clientes respondem pesquisas da mesma forma. Diferenças de perfil e comportamento podem influenciar a nota observada.

**5. Uso isolado da métrica**  
Analisar o NPS sem considerar métricas operacionais pode limitar a compreensão do problema.

<h3 style="color:#2E86C1;">
5.5 Formulação analítica do problema
</h3>

A partir do problema de negócio, uma formulação analítica possível é:

Como explicar e, futuramente, antecipar a nota de satisfação (`nps_score`) de um cliente a partir de variáveis operacionais disponíveis ao longo da jornada de compra?

Essa formulação faz sentido porque:

- parte diretamente da dor do negócio  
- define de forma clara o fenômeno de interesse  
- conecta a análise descritiva atual com uma possível evolução para modelos preditivos  
- mantém alinhamento entre a meta de negócio e o objetivo analítico
